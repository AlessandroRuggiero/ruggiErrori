```python
from math import e,cos,log,tan,sqrt,sin,exp
import pandas as pd
import numpy as np
from typing import Any,Dict,Callable,Tuple
from collections import OrderedDict
from tqdm import tqdm
```

```python
class Row:
    def __init__(self,a,b,f:Callable[[float],float]) -> None:
        self.a:float = a
        self.b:float = b
        self.f = f
        self.update()
    
    def update (self):
        self.m:float = (self.a + self.b) / 2
        self.fa = self.f (self.a)
        self.fb = self.f (self.b)
        self.fm = self.f (self.m)

    def to_dict (self) -> Dict[str,Any]:
        return {
            "a":self.a,
            "b":self.b,
            "f (a)":self.fa,
            "f (b)":self.fb,
            "m":self.m,
            "f (m)":self.fm
        }
```

```python
def bisezione (a:float,b:float,f:Callable[[float],float],end:int = 100) -> pd.DataFrame:
    row = Row(a,b,f)
    df = pd.DataFrame([row.to_dict()])
    df.columns = ["a","b","f (a)","f (b)","m","f (m)"]
    for _ in tqdm(range (end)):
        if row.fm > 0:
            row.b = row.m
        else :
            row.a = row.m
        row.update ()
        df = df.append([row.to_dict()],ignore_index=True)
        if row.fm == 0:
           break
        
    return df
```

```python
def g (x:float) -> float:
    return e**sin(log(x)) - 2 + x**x
def h (x:float) -> float:
    return abs(sin(x)/log(x)) + 2
def f (x:float):
    return g (x) - h(x)
```

```python
df = bisezione (1.5,2,end = 100,f = f)
```
